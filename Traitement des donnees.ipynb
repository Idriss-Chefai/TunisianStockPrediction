```python
import json
import pandas as pd
import matplotlib.pyplot as plt
tunisian_stock_market = 'Tunisian-Stock-Market.json'
with open(tunisian_stock_market, 'r') as json_file:
    
    json_data = json.load(json_file)

df = pd.DataFrame(json_data)

# Filter data for 'BNA' stock
bna_data = df[df['Stock_name'] == 'BNA']

# Convert 'Date' column to datetime format for proper plotting
bna_data['Date'] = pd.to_datetime(bna_data['Date'])

# Create separate plots for each variable by date
fig, axes = plt.subplots(nrows=4, ncols=1, figsize=(12, 15))

# Plot 'Adj_Close'
axes[0].plot(bna_data['Date'], bna_data['Adj_Close'], label='Adj_Close')
axes[0].set_title('BNA Stock - Adjusted Close Prices')
axes[0].set_xlabel('Date')
axes[0].set_ylabel('Adjusted Close Price')
axes[0].legend()

# Plot 'Adj_low'
axes[1].plot(bna_data['Date'], bna_data['Adj_low'], label='Adj_low')
axes[1].set_title('BNA Stock - Adjusted Low Prices')
axes[1].set_xlabel('Date')
axes[1].set_ylabel('Adjusted Low Price')
axes[1].legend()

# Plot 'Adj_High'
axes[2].plot(bna_data['Date'], bna_data['Adj_High'], label='Adj_High')
axes[2].set_title('BNA Stock - Adjusted High Prices')
axes[2].set_xlabel('Date')
axes[2].set_ylabel('Adjusted High Price')
axes[2].legend()

# Plot 'Volume'
axes[3].plot(bna_data['Date'], bna_data['Volume'], label='Volume')
axes[3].set_title('BNA Stock - Volume')
axes[3].set_xlabel('Date')
axes[3].set_ylabel('Volume')
axes[3].legend()

plt.tight_layout()
plt.show()
```

```python
import json
import pandas as pd
import matplotlib.pyplot as plt
tunisian_stock_market = 'Tunisian-Stock-Market.json'
with open(tunisian_stock_market, 'r') as json_file:
    
    json_data = json.load(json_file)

df = pd.DataFrame(json_data)

df['Date'] = pd.to_datetime(df['Date'])

# Filter data for the years 2020 and 2021
filtered_data = df[df['Date'].dt.year.isin([2020, 2021])]

# Filter data for 'BNA' stock
bna_data = filtered_data[filtered_data['Stock_name'] == 'BNA']

# Create separate plots for each variable by date
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(15, 15))

# Plot 'Adj_Close'
axes[0,0].plot(bna_data['Date'], bna_data['Adj_Close'], label='Adj_Close')
axes[0,0].set_title('BNA Stock - Adjusted Close Prices')
axes[0,0].set_xlabel('Date')
axes[0,0].set_ylabel('Adjusted Close Price')
axes[0,0].legend()

# Plot 'Adj_low'
axes[0,1].plot(bna_data['Date'], bna_data['Adj_low'], label='Adj_low')
axes[0,1].set_title('BNA Stock - Adjusted Low Prices')
axes[0,1].set_xlabel('Date')
axes[0,1].set_ylabel('Adjusted Low Price')
axes[0,1].legend()

# Plot 'Adj_High'
axes[1,0].plot(bna_data['Date'], bna_data['Adj_High'], label='Adj_High')
axes[1,0].set_title('BNA Stock - Adjusted High Prices')
axes[1,0].set_xlabel('Date')
axes[1,0].set_ylabel('Adjusted High Price')
axes[1,0].legend()

# Plot 'Volume'
axes[1,1].plot(bna_data['Date'], bna_data['Volume'], label='Volume')
axes[1,1].set_title('BNA Stock - Volume')
axes[1,1].set_xlabel('Date')
axes[1,1].set_ylabel('Volume')
axes[1,1].legend()

plt.tight_layout()
plt.show()
```

```python
import json
import pandas as pd
import matplotlib.pyplot as plt
tunisian_stock_market = 'Tunisian-Stock-Market.json'
with open(tunisian_stock_market, 'r') as json_file:
    
    json_data = json.load(json_file)

df = pd.DataFrame(json_data)

df['Date'] = pd.to_datetime(df['Date'])

# Filter data for the years 2020 and 2021
filtered_data = df[df['Date'].dt.year.isin([2020, 2021])]

# Filter data for 'BNA' stock
bna_data = filtered_data[filtered_data['Stock_name'] == 'BNA']

# Create a single plot for all variables by date
fig, ax = plt.subplots(figsize=(15, 8))

# Plot 'Adj_Close' Every 10 days
ax.plot(bna_data['Date'], bna_data['Adj_Close'].rolling(10).mean(), label='Adj_Close-10')
# Plot 'Adj_Close' Every 10 days
ax.plot(bna_data['Date'], bna_data['Adj_Close'].rolling(20).mean(), label='Adj_Close-20')
# Plot 'Adj_Close' Every 10 days
ax.plot(bna_data['Date'], bna_data['Adj_Close'].rolling(30).mean(), label='Adj_Close-50')
# Plot 'Adj_Close' Raw
ax.plot(bna_data['Date'], bna_data['Adj_Close'], label='Adj_Close')

ax.set_title('BNA Stock - Adjusted Prices')
ax.set_xlabel('Date')
ax.set_ylabel('Adjusted Prices')
ax.legend()

# Set x-axis ticks for better readability
ax.xaxis.set_major_locator(plt.MaxNLocator(nbins=10))

plt.show()
```
